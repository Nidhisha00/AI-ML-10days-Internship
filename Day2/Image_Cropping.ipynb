```python
# img[row_start:row_end, column_start:column_end]
# Height = 324 pixels
# Width  = 640 pixels

# The slicing indices for cropping were incorrect.
# In Python slicing, the start index must be less than the end index.
# For example, to crop from row 40 to 300 and column 390 to 640:
crop = img[40:300, 390:640]
cv2_imshow(crop)
```

```python
crop = img[40:300, 0:250]
cv2_imshow(crop)
```

```python
# Activity
crop = img[40:300, 200:450]
cv2_imshow(crop)
```
