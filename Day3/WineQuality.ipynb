```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/wine_quality_merged.csv')
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['quality_mean_imputed']=df['quality'].fillna(df['quality'].mean())
```

```python
df['quality_median_imputed']=df['quality'].fillna(df['quality'].median())
```

```python
import seaborn as sns
sns.set(style='whitegrid')
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='quality',y='alcohol',palette='pastel')

plt.title('Boxplot of  quality',fontsize=16)
plt.xlabel('quality',fontsize=12)
plt.ylabel('alcohol',fontsize=12)
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x='quality',y='alcohol',palette='pastel')

plt.title('lineplot of  quality',fontsize=16)
plt.xlabel('quality',fontsize=12)
plt.ylabel('alcohol',fontsize=12)
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='quality',y='alcohol',palette='pastel')

plt.title('Histogram of  quality',fontsize=16)
plt.xlabel('quality',fontsize=12)
plt.ylabel('alcohol',fontsize=12)
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='alcohol',kde=True)

plt.title('Histogram of  alcohol',fontsize=16)
plt.xlabel('alcohol',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.xlabel('') # Clear x-label as it's not applicable for correlation matrix
plt.ylabel('') # Clear y-label as it's not applicable for correlation matrix
plt.show()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

# Get counts for the 'type' column to create a pie chart
type_counts = df['type'].value_counts()

plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribution of Wine Type',fontsize=16)
plt.xlabel('') # Pie charts don't typically have x-labels
plt.ylabel('') # Pie charts don't typically have y-labels
plt.show()
```
