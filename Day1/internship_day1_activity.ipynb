```python
from sklearn.preprocessing import LabelEncoder
encoded_data=df.copy()

# Identify categorical columns for label encoding
categorical_cols = ['Sex', 'Embarked']

# Apply LabelEncoder to each categorical column
for col in categorical_cols:
    le = LabelEncoder()
    encoded_data[col] = le.fit_transform(encoded_data[col])

print(encoded_data[categorical_cols].head())
```

sklearn=>module................
prerocessing=>library in sklearn for label encoding................
label encoding is done in this


```python
label_encode=LabelEncoder()
encoded_data['Sex']=label_encode.fit_transform(encoded_data['Sex'])
```


label encoding for 'Sex' feature

```python
one_hot_encoded = pd.get_dummies(encoded_data['Embarked'], prefix='Embarked')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

# Since 'Pclass' is ordinal, we can create a mapped column for demonstration.
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
# 3. Ordinal Encoding For p class
encoded_data['Pclass_Category'] = encoded_data['Pclass'].map(pclass_mapping)

# Display the resulting dataset with relevant encoded columns
print(encoded_data[['Sex', 'Embarked', 'Embarked_0', 'Embarked_1', 'Embarked_2', 'Pclass', 'Pclass_Category']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler
```

```python
file_path='/content/titanic uncleaned.csv'
titanic_data=pd.read_csv(file_path)
```

```python
numerical_columns=['Age','Fare','Pclass']
```

```python
titanic_data.dropna(subset=numerical_columns,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
robust_scaler=RobustScaler()

```

```python
titanic_data['Age_MinMax']=minmax_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_MinMax']=minmax_scaler.fit_transform(titanic_data[['Fare']])
```

```python
titanic_data['Age_Standard']=standard_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_Standard']=standard_scaler.fit_transform(titanic_data[['Fare']])
```

```python
titanic_data['Age_Robust']=robust_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_Robust']=robust_scaler.fit_transform(titanic_data[['Fare']])
```

```python
titanic_data[['Age','Age_MinMax','Age_Standard','Age_Robust']]
```
