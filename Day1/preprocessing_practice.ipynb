```python
encoded_data = df.copy()
one_hot_encoded = pd.get_dummies(encoded_data['Attrition'], prefix='Attrition')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

# Since 'Pclass' is ordinal, we can create a mapped column for demonstration.
performance_rating_mapping = {1: 'First', 2: 'Second', 3: 'Third',4:'Fourth',5:'Fifth'}
# 3. Ordinal Encoding For p class
encoded_data['Performance_Rating_Category'] = encoded_data['Performance_Rating'].map(performance_rating_mapping)

# Display the resulting dataset with relevant encoded columns
print(encoded_data[['Age','Attrition','Attrition_Yes','Attrition_No','Performance_Rating','Performance_Rating_Category']].head())
```

```python
file_path='/content/preprocessing_practice_dataset.csv'
```

```python
df=pd.read_csv(file_path)
```

```python
numerical_columns=['Age','Experience_Years','Monthly_Salary']
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
robust_scaler=RobustScaler()
```

```python
df['Age_MinMax']=minmax_scaler.fit_transform(df[['Age']])
df['Experience_Years_MinMax']=minmax_scaler.fit_transform(df[['Experience_Years']])
df['Monthly_Salary_MinMax']=minmax_scaler.fit_transform(df[['Monthly_Salary']])
```

```python
df['Age_Standard']=standard_scaler.fit_transform(df[['Age']])
df['Experience_Years_Standard']=standard_scaler.fit_transform(df[['Experience_Years']])
df['Monthly_Salary_Standard']=standard_scaler.fit_transform(df[['Monthly_Salary']])
```

```python
df['Age_Robust']=robust_scaler.fit_transform(df[['Age']])
df['Experience_Years_Robust']=robust_scaler.fit_transform(df[['Experience_Years']])
df['Monthly_Salary_Robust']=robust_scaler.fit_transform(df[['Monthly_Salary']])
```
